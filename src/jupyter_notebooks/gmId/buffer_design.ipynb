```python
import numpy as np
from pygmid import Lookup
from pygmid import Lookup as lk
import pandas as pd
```

```python
n = lk('Chipathon2025/lookup_tables/nfet_03v3.mat')
p = lk('Chipathon2025/lookup_tables/pfet_03v3.mat')
```

n = lk('Chipathon2025/lookbup_tables/nfet_03v3.mat')
p = lk('Chipathon2025/lookup_tables/pfet_03v3.mat')

```python
## Sizing for Source follower circuit
gm = 62.8e-3 # based on calculation 
gm_id = np.array([5, 8, 10, 20])

l = 0.28

id = gm/gm_id
jd = n.lookup('ID_W', GM_ID=gm_id, L=l)
w = id/jd

cgg_w = n.lookup('CGG_W', GM_ID=gm_id, L=l)
cgg = w*cgg_w
ft = gm/cgg/2/np.pi

df = pd.DataFrame([gm_id, id, jd, w, cgg, ft], ['gm_id', 'id', 'jd', 'w', 'cgg', 'ft'], columns=[f'gm-id: {gm_id[i]}' for i in range(len(gm_id))]); 
df

## Suggestion is to go for gm-id = 8 # please help validate
```

```python
# Sizing for nfets in OTA
gm_ota = 210e-6 # from calculation

l = 0.28
gm_id = np.array([5, 8, 10, 15, 20])
id_ota = gm_ota/gm_id
jd = n.lookup('ID_W', GM_ID=gm_id, L=l)
w = id_ota/jd

cgg_w = n.lookup('CGG_W', GM_ID=gm_id, L=l)
cgg = w*cgg_w
ft = gm_ota/cgg/2/np.pi

df = pd.DataFrame([gm_id, id_ota, jd, w, cgg, ft], ['gm_id', 'id', 'jd', 'w', 'cgg', 'ft'], columns=[f'gm-id: {gm_id[i]}' for i in range(len(gm_id))]); 
df

## Suggestion is to go for gm-id = 15 # please help validate
```
